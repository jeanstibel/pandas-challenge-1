## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
sub_data_df = data_df[['unit_price', 'qty']].copy()
data_df['line_subtotal'] = sub_data_df['unit_price'] * sub_data_df['qty']
data_df[['unit_price', 'qty', 'line_subtotal']].head(2)
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 
# per pound for items 50 pounds or under.
total_weight_df = data_df[['unit_weight', 'qty']].copy()
data_df['total_weight'] = total_weight_df['unit_weight'] * total_weight_df['qty']

def shipping_price(row):
    if row['total_weight'] > 50:
        return row['total_weight'] * 7
    else:
        return row['total_weight'] * 10
    
data_df['shipping_price'] = data_df.apply(shipping_price, axis=1)
data_df[['unit_price', 'unit_weight', 'qty', 'total_weight', 'shipping_price']].head(3)
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
line_price = data_df[['line_subtotal', 'shipping_price']]
data_df['line_price'] = data_df['line_subtotal'] + data_df['shipping_price'] + (data_df['line_subtotal'] + data_df['shipping_price']) * 0.0925
data_df[['line_subtotal', 'shipping_price', 'line_price']].round(2).head(3)
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).

line_cost = data_df[['unit_cost', 'qty', 'shipping_price']].copy()
data_df['line_cost'] = line_cost['unit_cost'] * line_cost['qty'] + line_cost['shipping_price']
data_df.head(3)
```

```python
# Create a column for the profit of each line using line cost and line price

line_profit = data_df[['line_subtotal', 'total_weight', 'shipping_price']]

data_df['line_price'] = data_df['line_subtotal'] + data_df['shipping_price']\
     + (data_df['line_subtotal'] + data_df['shipping_price']) * 0.0925

data_df['line_profit'] = data_df['line_price'] - data_df['line_cost']
data_df.round(2).head(3)
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? 
# Check your work from Part 1 for client ids.
top_5_clients_df = data_df['client_id'].value_counts()

top_5_clients_total = data_df[data_df['client_id'].isin(top_5_clients_list)].groupby('client_id')['line_price'].sum()

for client_id, total_spent in top_5_clients_total.items():
    print(f"Client {client_id}: ${total_spent:.2f}")
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 

top_5_clients_total = data_df[data_df['client_id'].isin(top_5_clients_list)].groupby('client_id')\
    [['qty','shipping_price','line_price', 'line_cost', 'line_profit']].sum().round(2)

top_5_clients_total.head(5)
```

```python
# Format the data and rename the columns to names suitable for presentation.
data_df.rename(columns={
    'first': 'First Name',
    'last': 'Last Name',
    'job': 'Job',
    'phone': 'Phone',
    'email': 'Email',
    'client_id': 'Client ID',
    'order_id': 'Order ID',
    'order_date': 'Order Date',
    'order_week': 'Order Week',
    'order_year': 'Order Year',
    'item_id': 'Item',
    'category': 'Category',
    'subcategory': 'Subcategory',
    'unit_price': 'Unit Price',
    'unit_cost': 'Unit Cost',
    'unit_weight': 'Unit Weight',
    'qty': 'Units',
    'line_number': 'Line Number',
    'line_subtotal': 'Line Subtotal',
    'total_weight': 'Total Weight',
    'shipping_price': 'Shipping Price',
    'line_price': 'Total Price',
    'line_cost': 'Total Cost',
    'line_profit': 'Total Profit'
}, inplace=True)

data_df.head(5) 
```

```python
# Define the money columns. 
money_columns = ['Shipping Price', 'Total Price', 'Total Cost',
                 'Total Profit']
money_columns
```

```python
### Define a function that converts a dollar amount to millions.
def convert_to_millions(money_columns):
    return money_columns / 1000000
```

```python
# Apply the currency_format_millions function to only the money columns.
# Rename the columns to reflect the change in the money format. 
data_df.rename(columns={
    'Shipping Price': 'Shipping Price (millions)', 
    'Total Price': 'Total Price (millions)', 
    'Total Cost': 'Total Cost (millions)', 
    'Total Profit': 'Total Profit (millions)'
}, inplace=True)

data_df['Shipping Price (millions)'] = convert_to_millions(data_df['Shipping Price (millions)'])
data_df['Total Price (millions)'] = convert_to_millions(data_df['Total Price (millions)'])
data_df['Total Cost (millions)'] = convert_to_millions(data_df['Total Cost (millions)'])
data_df['Total Profit (millions)'] = convert_to_millions(data_df['Total Profit (millions)'])

top_5_clients = data_df[data_df['Client ID'].isin(top_5_clients_list)][['Client ID', 'Units', 'Shipping Price (millions)', 'Total Cost (millions)', 'Total Price (millions)', 'Total Profit (millions)']]

top_5_clients_total = top_5_clients.groupby('Client ID').sum()
top_5_clients_total
```

```python
### Sort the updated data by "Total Profit (millions)" from highest to lowest and 
### assign the sort to a new DataFrame.
top_5_clients_total_sorted = top_5_clients.groupby('Client ID').sum()
top_5_clients_total_sorted_df = top_5_clients_total_sorted.sort_values(by="Total Profit (millions)", ascending=False)
top_5_clients_total_sorted_df
```
